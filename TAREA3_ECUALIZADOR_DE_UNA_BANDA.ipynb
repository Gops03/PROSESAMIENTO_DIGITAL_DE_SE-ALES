```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Par√°metros de la se√±al
f_start = 10
f_end = 500
f_step = 10
step_duration = 0.5  # segundos
fs = 5000

num_steps = (f_end - f_start) // f_step + 1
total_duration = num_steps * step_duration
t = np.linspace(0, total_duration, int(fs * total_duration), endpoint=False)

# Generar la se√±al
signal = np.zeros_like(t)
for i in range(num_steps):
    freq = f_start + i * f_step
    t_start = i * step_duration
    t_end = (i + 1) * step_duration
    idx = (t >= t_start) & (t < t_end)
    signal[idx] = np.sin(2 * np.pi * freq * t[idx])

# Funci√≥n para graficar un segmento
def plot_segment(inicio_simulacion):
    start_idx = int(inicio_simulacion * fs)
    end_idx = start_idx + int(5 * fs)
    if end_idx > len(t):
        end_idx = len(t)
    plt.figure(figsize=(12, 4))
    plt.plot(t[start_idx:end_idx], signal[start_idx:end_idx])
    plt.title(f"Se√±al desde {inicio_simulacion} s hasta {inicio_simulacion + 5} s")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid()
    plt.show()

# Slider con etiqueta personalizada
slider = widgets.IntSlider(
    value=0,
    min=0,
    max=total_duration - 5,
    step=1,
    description='Inicio de simulaci√≥n:',
    continuous_update=False
)

interact(plot_segment, inicio_simulacion=slider);


```

```python
from scipy.signal import iirnotch, filtfilt


# Funci√≥n para aplicar el filtro notch y graficar la se√±al filtrada
def bandstop_filter_plot(freq_rechazo, Q):
    # Crear filtro notch
    b, a = iirnotch(w0=freq_rechazo, Q=Q, fs=fs)
    signal_filtrada = filtfilt(b, a, signal)

    # Graficar se√±al original y filtrada (mismo segmento)
    start_idx = 0
    end_idx = int(5 * fs)

    plt.figure(figsize=(12, 5))
    plt.plot(t[start_idx:end_idx], signal[start_idx:end_idx], label='Original', alpha=0.5)
    plt.plot(t[start_idx:end_idx], signal_filtrada[start_idx:end_idx], label='Filtrada (Notch)', color='red')
    plt.title(f"Filtro Rechaza Banda: {freq_rechazo} Hz (Q={Q})")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

# Sliders interactivos para frecuencia y Q
freq_slider = widgets.FloatSlider(
    value=60,
    min=10,
    max=500,
    step=1,
    description='Frecuencia:',
    continuous_update=False
)

Q_slider = widgets.FloatSlider(
    value=30.0,
    min=1.0,
    max=100.0,
    step=1.0,
    description='Q (ancho):',
    continuous_update=False
)

# Interactividad
interact(bandstop_filter_plot, freq_rechazo=freq_slider, Q=Q_slider);
```
